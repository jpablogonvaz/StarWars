```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("starwars.csv")
```

```python
df = df.drop(df.iloc[:,38:],1)
```

```python
df = df.rename(columns= {'Which of the following Star Wars films have you seen? Please select all that apply.'
                          : 'Have you seen Episode I The Phantom Menace',
                          'Unnamed: 4' : 'Have you seen Episode II Attack of the Clones',
                          'Unnamed: 5' : 'Have you seen Episode III Revenge of the Sith',
                          'Unnamed: 6' : 'Have you seen Episode IV A New Hope',
                          'Unnamed: 7' : 'Have you seen Episode V The Empire Strikes Back',
                          'Unnamed: 8' : 'Have you seen Episode VI Return of the Jedi'})
```

```python
df = df.rename(columns= {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.'
                          : 'Ranking Episode I The Phantom Menace',
                          'Unnamed: 10' : 'Ranking Episode II Attack of the Clones',
                          'Unnamed: 11' : 'Ranking Episode III Revenge of the Sith',
                          'Unnamed: 12' : 'Ranking Episode IV A New Hope',
                          'Unnamed: 13' : 'Ranking Episode V The Empire Strikes Back',
                          'Unnamed: 14' : 'Ranking Episode VI Return of the Jedi from'})
```

```python
df = df.rename(columns= {'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.'
                          : 'Han Solo',
                          'Unnamed: 16' : 'Luke Skywalker',
                          'Unnamed: 17' : 'Princess Leia Organa',
                          'Unnamed: 18' : 'Anakin Skywalker',
                          'Unnamed: 19' : 'Obi Wan Kenobi',
                          'Unnamed: 20' : 'Emperor Palpatine',
                          'Unnamed: 21' : 'Darth Vader',
                          'Unnamed: 22' : 'Lando Calrissian',
                          'Unnamed: 23' : 'Boba Fett',
                          'Unnamed: 24' : 'C-3P0',
                          'Unnamed: 25' : 'R2 D2',
                          'Unnamed: 26' : 'Jar Jar Binks',
                          'Unnamed: 27' : 'Padme Amidala',
                          'Unnamed: 28' : 'Yoda',})
```

```python
df
```

```python
df = df.replace(['Very unfavorably',
                 'Somewhat unfavorably',
                 'Unfamiliar (N/A)',
                 'Neither favorably nor unfavorably (neutral)',
                 'Somewhat favorably',
                 'Very favorably'], 
                     [-2, -1, 0, 0, 1, 2])
df['Han Solo'] = df['Han Solo'].fillna(0)
df['Luke Skywalker'] = df['Luke Skywalker'].fillna(0)
df['Princess Leia Organa'] = df['Princess Leia Organa'].fillna(0)
df['Anakin Skywalker'] = df['Anakin Skywalker'].fillna(0)
df['Obi Wan Kenobi'] = df['Obi Wan Kenobi'].fillna(0)
df['Emperor Palpatine'] = df['Emperor Palpatine'].fillna(0)
df['Darth Vader'] = df['Darth Vader'].fillna(0)
df['Lando Calrissian'] = df['Lando Calrissian'].fillna(0)
df['Boba Fett'] = df['Boba Fett'].fillna(0)
df['C-3P0'] = df['C-3P0'].fillna(0)
df['R2 D2'] = df['R2 D2'].fillna(0)
df['Jar Jar Binks'] = df['Jar Jar Binks'].fillna(0)
df['Padme Amidala'] = df['Padme Amidala'].fillna(0)
df['Yoda'] = df['Yoda'].fillna(0)
```

```python
df = df.replace(['Star Wars: Episode I  The Phantom Menace',
                'Star Wars: Episode II  Attack of the Clones',
                 'Star Wars: Episode III  Revenge of the Sith',
                'Star Wars: Episode IV  A New Hope',
                'Star Wars: Episode V The Empire Strikes Back',
                'Star Wars: Episode VI Return of the Jedi'],[1,1,1,1,1,1])
df['Have you seen Episode I The Phantom Menace'] = df['Have you seen Episode I The Phantom Menace'].fillna(-1)
df['Have you seen Episode II Attack of the Clones'] = df['Have you seen Episode II Attack of the Clones'].fillna(-1)
df['Have you seen Episode III Revenge of the Sith'] = df['Have you seen Episode III Revenge of the Sith'].fillna(-1)
df['Have you seen Episode IV A New Hope'] = df['Have you seen Episode IV A New Hope'].fillna(-1)
df['Have you seen Episode V The Empire Strikes Back'] = df['Have you seen Episode V The Empire Strikes Back'].fillna(-1)
df['Have you seen Episode VI Return of the Jedi'] = df['Have you seen Episode VI Return of the Jedi'].fillna(-1)
```

```python
df = df.replace(['Yes', 'No'],[1,-1])
df['Have you seen any of the 6 films in the Star Wars franchise?'] = df['Have you seen any of the 6 films in the Star Wars franchise?'].fillna(0)
df['Do you consider yourself to be a fan of the Star Wars film franchise?'] = df['Do you consider yourself to be a fan of the Star Wars film franchise?'].fillna(0)
df['Are you familiar with the Expanded Universe?'] = df['Are you familiar with the Expanded Universe?'].fillna(0)
df['Do you consider yourself to be a fan of the Expanded Universe?'] = df['Do you consider yourself to be a fan of the Expanded Universe?'].fillna(0)
df['Do you consider yourself to be a fan of the Star Trek franchise?'] = df['Do you consider yourself to be a fan of the Star Trek franchise?'].fillna(0)
```

```python
df['Ranking Episode I The Phantom Menace'] = df['Ranking Episode I The Phantom Menace'].fillna(0)
df['Ranking Episode II Attack of the Clones'] = df['Ranking Episode II Attack of the Clones'].fillna(0)
df['Ranking Episode III Revenge of the Sith'] = df['Ranking Episode III Revenge of the Sith'].fillna(0)
df['Ranking Episode IV A New Hope'] = df['Ranking Episode IV A New Hope'].fillna(0)
df['Ranking Episode V The Empire Strikes Back'] = df['Ranking Episode V The Empire Strikes Back'].fillna(0)
df['Ranking Episode VI Return of the Jedi from'] = df['Ranking Episode VI Return of the Jedi from'].fillna(0)
```

```python
df = df.replace(['$0 - $24,999', '$25,000 - $49,999',
                 '$50,000 - $99,999', '$100,000 - $149,999','$150,000+'],[0,1,2,3,4])
```

```python
df = df.replace(['Less than high school degree', 'High school degree', 
                 'Some college or Associate degree', 'Bachelor degree',  
                 'Graduate degree', ],[0,1,2,3,4])
```

```python
df = df.replace(['Male', 'Female'],[1,0])
```

```python
df = df.replace(['18-29', '30-44',
                 '45-60', '> 60'],[0,1,2,3])
```

```python
df = df.replace(["I don't understand this question", "Greedo",
                 'Han'],[-1,1,2])
```

```python
df['fan'] = df['Do you consider yourself to be a fan of the Star Wars film franchise?'] + \
df['Are you familiar with the Expanded Universe?'] + \
df['Do you consider yourself to be a fan of the Expanded Universe?'] + \
df['Do you consider yourself to be a fan of the Star Trek franchise?'] + \
df['Which character shot first?'] + \
df['Have you seen Episode I The Phantom Menace'] + \
df['Have you seen Episode II Attack of the Clones'] + \
df['Have you seen Episode III Revenge of the Sith'] + \
df['Have you seen Episode IV A New Hope'] + \
df['Have you seen Episode V The Empire Strikes Back'] + \
df['Have you seen Episode VI Return of the Jedi']

#df['fan'] = df['fan']/11
```
